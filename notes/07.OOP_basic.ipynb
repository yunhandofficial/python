````python
# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class MyList:
    data = []
    def append(self, el):
        self.data +=[el]
        
    def pop(self):
        a=self.data[-1]
        self.data = self.data[:-1]
        return a
    
    def reverse(self):
        self.data=self.data[::-1]
        
    def count(self,el):
        return self.data.count(el)
    
    def clear(self):
        self.data=[]
        
    def __repr__(self):
        return f'ë‚´ ë¦¬ìŠ¤íŠ¸ì—ëŠ” {self.data}ì´ ë‹´ê²¨ìˆë‹¤.'
````
